**Retrieval Augmented Generation (RAG):** Describes the process of implementing semantic similarity processing into large language model context pipelines so that large language models are able to be systematically fed the most accurate, relevant information possible when prompted.

**Chunking:** The process by which we separate documents into groups of sentences (chunks) based on semantic similarity. Think of breaking up a document into "paragraphs."

## 1.0 Foundation - Understanding the Problem

When we use natural language processing to talk to LLM's, we need to ensure we give them semantically relevant, accurate information. We also want to make sure we don't give them *too* much information as well. This is why we build RAG systems that allow us to retrieve relevant information so that a model can give an accurate output based on the user's input prompt.

Let's say a user talks to an LLM about a question inside a document. A typical LLM pipeline might look like this:

0. Prior to any user interaction, the document would be chunked according to semantic similarity, and these chunks of text would be stored in a "vector store" for retrieval upon prompting (explained later).
1. User inputs a prompt into an LLM.
2. User's prompt is first directly fed to an embedding model.
3. Embedding is sent to the vector store service that hosts the document's text and embeddings, almost like a "library lookup" for relevant text.
4. Most semantically relevant text from the document is returned back to the model as additional context, along with the original text prompt.
5. LLM is able to answer the question accurately and reliably without becoming overloaded.

## 1.2 Finding The "Best" Chunks

Going back to Harry Potter, imagine we could somehow calculate the similarity between sentences within the same paragraph. Ideally, they should be very similar. Now imagine we could calculate the similarity between random sentences from different paragraphs. Ideally, despite being random, they would be *more dissimilar than similar*. This is the foundational understanding behind the Chunking subnet's reward heuristic.

### The Reward Heuristic
The Bittensor Chunking subnet used a mathematical way to measure chunk quality. Here's the simplified version:

```python
import numpy as np
from typing import List

def simple_reward_score(chunks: List[List[str]], embeddings: np.ndarray) -> float:
    """
    Simplified version of the reward heuristic.
    
    The idea: Good chunks have high internal similarity, 
    low external similarity between different chunks.
    """
    
    # Step 1: Calculate average similarity WITHIN each chunk
    intra_chunk_similarities = []
    
    for chunk_sentences in chunks:
        if len(chunk_sentences) < 2:
            continue
            
        # Get similarities between all pairs within this chunk
        chunk_sims = []
        for i in range(len(chunk_sentences)):
            for j in range(i + 1, len(chunk_sentences)):
                # In real code, we'd use actual embeddings
                # Here we'll simulate similarity
                similarity = np.random.random()  # Placeholder
                chunk_sims.append(similarity)
        
        if chunk_sims:
            intra_chunk_similarities.append(np.mean(chunk_sims))
    
    # Step 2: Calculate average similarity BETWEEN different chunks  
    inter_chunk_similarities = []
    
    # Compare sentences from different chunks
    for i, chunk1 in enumerate(chunks):
        for j, chunk2 in enumerate(chunks):
            if i >= j:  # Don't double-count
                continue
                
            # Compare sentences between these two chunks
            between_sims = []
            for sent1 in chunk1:
                for sent2 in chunk2:
                    similarity = np.random.random()  # Placeholder
                    between_sims.append(similarity)
            
            if between_sims:
                inter_chunk_similarities.append(np.mean(between_sims))
    
    # Step 3: The reward is the ratio
    avg_intra = np.mean(intra_chunk_similarities) if intra_chunk_similarities else 0
    avg_inter = np.mean(inter_chunk_similarities) if inter_chunk_similarities else 0
    
    # Good chunks: high internal similarity, low external similarity
    if avg_inter > 0:
        reward = avg_intra / avg_inter
    else:
        reward = avg_intra
        
    return reward

# Test it out
chunks_example = [
    ["Harry Potter was a wizard", "He wore round glasses"],
    ["Climate change is serious", "Sea levels are rising"]
]

score = simple_reward_score(chunks_example, None)  # Embeddings placeholder
print(f"Chunk quality score: {score:.3f}")
print("\nHigher scores = better chunks!")
```

This method of scoring chunks typically results in a score of `1.01` or greater, and requires all chunks in the document to be within a max chunk size and chunk quantity. By maximizing the difference between intrachunk and interchunk similarity across many random samples, algorithms highly optimized for this heuristic gradually produced better scores over time.

## 2.0 Problems With Basic Chunking

Basic chunking methods, while fast and effective for many documents, have several problems. Take a look below to see an example of size-based chunking, which doesn't incorporate semantic similarity at all:

```python
import sys
import os
sys.path.append('research/building_blocks')
sys.path.append('research/visualizations')

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from typing import List, Dict, Tuple, Optional
import warnings
warnings.filterwarnings('ignore')

# Configure plotting
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

def naive_recursive_chunking(text: str, max_size: int) -> List[str]:
    """
    Simple size-based chunking that ignores semantic boundaries.
    This is what most systems do - and why it fails for RAG.
    """
    words = text.split()
    chunks = []
    current_chunk = []
    current_size = 0
    
    for word in words:
        word_size = len(word) + 1  # +1 for space
        if current_size + word_size > max_size and current_chunk:
            chunks.append(' '.join(current_chunk))
            current_chunk = [word]
            current_size = word_size
        else:
            current_chunk.append(word)
            current_size += word_size
    
    if current_chunk:
        chunks.append(' '.join(current_chunk))
    
    return chunks

# Demonstrate the problem with naive chunking
sample_text = """
Artificial intelligence has revolutionized many fields. Machine learning algorithms can now 
process vast amounts of data to identify patterns. Deep learning networks use multiple layers 
to extract features from raw input. Natural language processing enables computers to understand 
human text. Computer vision allows machines to interpret visual information. These technologies 
are transforming industries worldwide.
"""

print("Sample text:")
print(f'"{sample_text.strip()}"')
print()

naive_chunks = naive_recursive_chunking(sample_text.strip(), max_size=100)
print("Chunks based on max_size:")
for i, chunk in enumerate(naive_chunks):
    print(f"  Chunk {i+1}: {chunk}")
print()
```

## 3.0 The Similarity Matrix

Back to Harry Potter. No matter how you might chunk the book into paragraphs, each sentence means something. What if we could compare every single sentence to every other sentence and store this data in an array? We can do this using a "similarity matrix:"

```python
print("🔧 Section 2: Building Block 1 - Similarity Matrices")
print("-" * 55)

# Import our similarity matrix building block
from similarity_matrices import SimilarityMatrixBuilder, SimilarityMatrixConfig

print("Core Innovation: Multi-Window Similarity Matrices")
print("=" * 50)
print()

print("Traditional approach:")
print("  • Compare sentence A with sentence B")
print("  • Single similarity score")
print("  • Limited context")
print()

print("Our approach:")
print("  • Compare windows of 1, 2, and 3 sentences")
print("  • Captures broader semantic context")
print("  • Enhanced with mathematical techniques:")
print("    - Edge detection for boundary identification")
print("    - Block measure analysis for coherent regions")
print("    - Advanced normalization methods")
print()

# Create a more detailed sample document for demonstration
sample_sentences = [
    "Artificial intelligence has transformed modern computing.",
    "Machine learning algorithms can identify complex patterns in data.",
    "Deep learning uses neural networks with multiple hidden layers.",
    "Natural language processing enables computers to understand text.",
    "Computer vision allows machines to interpret visual information.",
    "These AI technologies are revolutionizing many industries.",
    "Climate change poses significant challenges to our planet.",
    "Rising temperatures are melting polar ice caps rapidly.",
    "Extreme weather events are becoming more frequent globally.",
    "Renewable energy sources offer sustainable alternatives.",
    "Solar and wind power are increasingly cost-effective.",
    "Governments worldwide are implementing green policies."
]

print(f"📄 Sample document with {len(sample_sentences)} sentences:")
for i, sentence in enumerate(sample_sentences):
    print(f"  {i+1:2d}. {sentence}")
print()

# Create sample embeddings (in a real scenario, you'd use sentence-transformers)
def create_sample_embeddings(sentences: List[str], embedding_dim: int = 384) -> np.ndarray:
    """
    Create sample embeddings that reflect semantic similarity.
    In practice, you would use sentence-transformers or OpenAI embeddings.
    """
    np.random.seed(42)
    
    # Create topic-based embeddings
    # Sentences 0-5: AI/ML topic
    # Sentences 6-11: Climate change topic
    
    embeddings = []
    for i, sentence in enumerate(sentences):
        if i <= 5:  # AI/ML topic
            base_vector = np.array([1.0, 0.0, 0.5])  # AI topic signature
        else:  # Climate change topic
            base_vector = np.array([0.0, 1.0, 0.5])  # Climate topic signature
        
        # Add some noise and random components to make it realistic
        noise = np.random.normal(0, 0.1, embedding_dim - 3)
        full_embedding = np.concatenate([base_vector, noise])
        
        # Add some sentence-specific variation
        variation = np.random.normal(0, 0.05, embedding_dim)
        full_embedding += variation
        
        embeddings.append(full_embedding)
    
    return np.array(embeddings)

print("🧮 Step 1: Creating Basic Similarity Matrix")
print("-" * 42)

# Get embeddings for our sample sentences
embeddings = create_sample_embeddings(sample_sentences)
print(f"Created embeddings: {embeddings.shape}")

# Initialize the similarity matrix builder
config = SimilarityMatrixConfig(
    window_sizes=[1, 2, 3],
    enhancement_enabled=True,
    normalization_method="l2"
)

builder = SimilarityMatrixBuilder(config)

# Create basic similarity matrix
basic_matrix = builder.create_basic_similarity_matrix(embeddings)
print(f"Basic similarity matrix shape: {basic_matrix.shape}")
print(f"Similarity range: [{basic_matrix.min():.3f}, {basic_matrix.max():.3f}]")
print()

print("🎨 Visualizing the Basic Similarity Matrix:")

plt.figure(figsize=(10, 8))
sns.heatmap(basic_matrix, 
            annot=False, 
            cmap='viridis', 
            square=True,
            cbar_kws={'label': 'Cosine Similarity'})
plt.title('Basic Similarity Matrix\n(Darker = More Similar)')
plt.xlabel('Sentence Index')
plt.ylabel('Sentence Index')
plt.tight_layout()
plt.show()

print("📊 Key observations:")
print("  • Dark blocks along diagonal = high self-similarity")
print("  • Sentences 0-5 show higher similarity (AI topic)")
print("  • Sentences 6-11 show higher similarity (Climate topic)")
print("  • Clear transition around sentence 6-7")
print()

print("🔬 Step 2: Matrix Enhancement Techniques")
print("-" * 42)

# Apply enhancement techniques
enhanced_matrix, enhancement_data = builder.enhance_similarity_matrix(basic_matrix)

print(f"Enhanced matrix range: [{enhanced_matrix.min():.3f}, {enhanced_matrix.max():.3f}]")
print(f"Enhancement data available: {list(enhancement_data.keys())}")
print()

# Visualize the enhancement process
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Original matrix
sns.heatmap(basic_matrix, ax=axes[0,0], cmap='viridis', square=True, cbar=True)
axes[0,0].set_title('Original Similarity Matrix')

# Enhanced matrix
sns.heatmap(enhanced_matrix, ax=axes[0,1], cmap='viridis', square=True, cbar=True)
axes[0,1].set_title('Enhanced Similarity Matrix')

# Edge detection
if 'edge_magnitude' in enhancement_data:
    sns.heatmap(enhancement_data['edge_magnitude'], ax=axes[1,0], cmap='hot', square=True, cbar=True)
    axes[1,0].set_title('Edge Magnitude (Boundary Detection)')

# Block measure
if 'block_measure' in enhancement_data:
    sns.heatmap(enhancement_data['block_measure'], ax=axes[1,1], cmap='plasma', square=True, cbar=True)
    axes[1,1].set_title('Block Measure (Coherence Detection)')

plt.tight_layout()
plt.show()

print("🎯 Enhancement Results:")
print("  • Edge detection highlights potential chunk boundaries")
print("  • Block measure identifies coherent regions")
print("  • Enhanced matrix balances similarity with boundary detection")
print()

print("📈 Similarity Analysis:")
print("-" * 20)

# Analyze similarity patterns
def analyze_similarity_patterns(matrix: np.ndarray, sentences: List[str]) -> None:
    """Analyze and report similarity patterns in the matrix."""
    
    # Find the most similar sentence pairs
    similarity_pairs = []
    n = matrix.shape[0]
    
    for i in range(n):
        for j in range(i+1, n):
            similarity_pairs.append((i, j, matrix[i, j]))
    
    # Sort by similarity
    similarity_pairs.sort(key=lambda x: x[2], reverse=True)
    
    print("🔗 Top 5 most similar sentence pairs:")
    for i, (sent1, sent2, sim) in enumerate(similarity_pairs[:5]):
        print(f"  {i+1}. Sentences {sent1+1}-{sent2+1} (similarity: {sim:.3f})")
        print(f"     '{sentences[sent1][:50]}...'")
        print(f"     '{sentences[sent2][:50]}...'")
        print()
    
    # Identify potential chunk boundaries (low similarity between adjacent sentences)
    adjacent_similarities = []
    for i in range(n-1):
        adjacent_similarities.append((i, i+1, matrix[i, i+1]))
    
    adjacent_similarities.sort(key=lambda x: x[2])
    
    print("✂️  Potential chunk boundaries (lowest adjacent similarities):")
    for i, (sent1, sent2, sim) in enumerate(adjacent_similarities[:3]):
        print(f"  {i+1}. Between sentences {sent1+1} and {sent2+1} (similarity: {sim:.3f})")
        print(f"     Split here: '{sentences[sent1][:40]}...' | '{sentences[sent2][:40]}...'")
        print()

analyze_similarity_patterns(enhanced_matrix, sample_sentences)

print("🎉 Building Block 1 Complete!")
print("=" * 30)
print("We now have:")
print("  ✓ Basic similarity matrix creation")
print("  ✓ Matrix enhancement techniques")
print("  ✓ Boundary detection capabilities")
print("  ✓ Coherence measurement tools")
print()
print("Next: Building Block 2 - Distance Calculations")
print("Coming up: How to convert similarities to distances and detect natural split points")
```
